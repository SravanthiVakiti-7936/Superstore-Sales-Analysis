```python
import pandas as pd

data = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
data.head()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt

# Total Sales and Profit
total_sales = data['Sales'].sum()
total_profit = data['Profit'].sum()
print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
region_sales = data.groupby('Region')['Sales'].sum()
region_profit = data.groupby('Region')['Profit'].sum()

# Plotting Sales by Region
region_sales.plot(kind='bar', title='Sales by Region', figsize=(8,5))
plt.ylabel('Sales')
plt.show()

# Plotting Profit by Region
region_profit.plot(kind='bar', title='Profit by Region', figsize=(8,5), color='green')
plt.ylabel('Profit')
plt.show()
```

```python
category_sales = data.groupby('Category')['Sales'].sum()
category_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales by Category', figsize=(6,6))
plt.show()
```

```python
top_products = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', title='Top 10 Products by Sales', figsize=(10,5), color='orange')
plt.ylabel('Sales')
plt.show()
```

```python
monthly_sales = data.groupby(data['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales Trend', figsize=(10,5), marker='o')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.show()
```
